```python
class ShoppingCart:
    def __init__(self):
        self.productA = 20
        self.productB = 40
        self.productC = 50
        self.giftWrap = 1
        self.shipping = 5
        self.totalAmountOfA = 0
        self.totalAmountOfB = 0
        self.totalAmountOfC = 0

    def totalAmountOfProductA(self, quantityA):
        self.totalAmountOfA = quantityA * self.productA
        return self.totalAmountOfA

    def totalAmountOfProductB(self, quantityB):
        self.totalAmountOfB = quantityB * self.productB
        return self.totalAmountOfB

    def totalAmountOfProductC(self, quantityC):
        self.totalAmountOfC = quantityC * self.productC
        return self.totalAmountOfC
    #Discount Rule 1
    def flat10Discount(self, subTotal):
        flat_10_discount = 0
        if subTotal > 200:
            flat_10_discount = 10
        return flat_10_discount
    #Discount Rule 2
    def bulk5Discount(self, quantityA, quantityB, quantityC):
        bulk_5_discount = 0
        bulk_5_discount_A = 0
        bulk_5_discount_B = 0
        bulk_5_discount_C = 0
        if quantityA > 10:
            bulk_5_discount_A = (self.totalAmountOfProductA(quantityA) * 5) / 100
            if quantityB > 10:
                bulk_5_discount_B = (self.totalAmountOfProductB(quantityB) * 5) / 100
                if quantityC > 10:
                    bulk_5_discount_C = (self.totalAmountOfProductC(quantityC) * 5) / 100
            elif quantityC > 10:
                bulk_5_discount_C = (self.totalAmountOfProductC(quantityC) * 5) / 100
            else:
                bulk_5_discount_B = 0
                bulk_5_discount_C = 0
        bulk_5_discount = bulk_5_discount_A + bulk_5_discount_B + bulk_5_discount_C
        return bulk_5_discount
    #Discount Rule 3
    def bulk10Discount(self, subTotal, totalQuantity):
        bulk_10_discount = 0
        if totalQuantity > 20:
            bulk_10_discount = (subTotal * 10) / 100
        return bulk_10_discount
    #Discount Rule 4
    def tiered50Discount(self, totalQuantity, quantityA, quantityB, quantityC):
        tiered50DiscountA = 0
        tiered50DiscountB = 0
        tiered50DiscountC = 0
        tiered_50_Discount = 0
        if totalQuantity > 30:
            if self.totalAmountOfProductA(quantityA) > 15:
                tiered50DiscountA = (self.productA * 15) + (((quantityA - 15) * self.productA) * 50) / 100
            else:
                tiered50DiscountA = 0
            if self.totalAmountOfProductB(quantityB) > 15:
                tiered50DiscountB = (self.productB * 15) + (((quantityB - 15) * self.productB) * 50) / 100
            else:
                tiered50DiscountB = 0
            if self.totalAmountOfProductC(quantityC) > 15:
                tiered50DiscountC = (self.productC * 15) + (((quantityC - 15) * self.productC) * 50) / 100
            else:
                tiered50DiscountC = 0
        tiered_50_Discount = tiered50DiscountA + tiered50DiscountB + tiered50DiscountC
        return 0
    
    def purchase(self, quantityA, quantityB, quantityC, giftWrapUnits):
        totalQuantity = quantityA + quantityB + quantityC
        subTotal = (quantityA * self.productA) + (quantityB * self.productB) + (quantityC * self.productC)
        flat_10_discount = self.flat10Discount(subTotal)
        bulk_5_discount = self.bulk5Discount(quantityA, quantityB, quantityC)
        bulk_10_discount = self.bulk10Discount(subTotal, totalQuantity)
        tiered_50_Discount = self.tiered50Discount(totalQuantity, quantityA, quantityB, quantityC)
        
        print("Product Name      Quantity      Total Amount of Product")
        print("-----------------------------------------------------------")
        print("Product A            {}            {}".format(quantityA, self.totalAmountOfProductA(quantityA)))
        print("Product B            {}            {}".format(quantityB, self.totalAmountOfProductB(quantityB)))
        print("Product C            {}            {}".format(quantityC, self.totalAmountOfProductC(quantityC)))
        print("-----------------------------------------------------------")
        print("Total Quantity       {}".format(totalQuantity))
        print("Sub Total            ${}".format(subTotal))
        
        if flat_10_discount > bulk_5_discount and flat_10_discount > bulk_10_discount and flat_10_discount > tiered_50_Discount:
            print("Coupon - $10        -${}".format(flat_10_discount))
            appliedDiscount = flat_10_discount
        elif bulk_5_discount > flat_10_discount and bulk_5_discount > bulk_10_discount and bulk_5_discount > tiered_50_Discount:
            print("Coupon - 5%        -${}".format(bulk_5_discount))
            appliedDiscount = bulk_5_discount
        elif bulk_10_discount > flat_10_discount and bulk_10_discount > bulk_5_discount and bulk_10_discount > tiered_50_Discount:
            print("Coupon - 10%        -${}".format(bulk_10_discount))
            appliedDiscount = bulk_10_discount
        else:
            print("Coupon - 50%        -${}".format(tiered_50_Discount))
            appliedDiscount = tiered_50_Discount
        #Calculating Shipping Cost
        pkgCount = 0
        shippingFees = 0
        if totalQuantity > 10:
            while totalQuantity > 0:
                totalQuantity -= 10
                pkgCount += 1
            shippingFees = pkgCount * self.shipping
        else:
            shippingFees = self.shipping
        #Calculating Gift Wrapping Fees    
        giftWrapFees = giftWrapUnits * self.giftWrap 
        #Calculating Total Bill 
        totalBillAmount = (subTotal + shippingFees + giftWrapFees) - appliedDiscount    
        
        print("Shipping Fees        ${}".format(shippingFees))
        print("Wrapping Fees        ${}".format(giftWrapFees))
        print()
        print("Total                ${}".format(totalBillAmount))
        return 0

shc = ShoppingCart() #object creation
quantityA = int(input("Enter the quantity for Product A : "))
quantityB = int(input("Enter the quantity for Product B : "))
quantityC = int(input("Enter the quantity for Product C : "))
yesOrNo = input("You want to wrap your product as gift or not? (Press Yes or No) ")
if yesOrNo.lower() == "yes":
    giftWrapUnits = int(input("How many units you have to wrap? "))
else:
    giftWrapUnits = 0
shc.purchase(quantityA, quantityB, quantityC, giftWrapUnits)

```
